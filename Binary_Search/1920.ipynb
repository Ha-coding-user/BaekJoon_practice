# 1920 수 찾기
다른 블로그 참조 : X

### Q)
N개의 정수 A[1], A[2], …, A[N]이 주어져 있을 때, 이 안에 X라는 정수가 존재하는지 알아내는 프로그램을 작성하시오.

### Input)
첫째 줄에 자연수 N(1 ≤ N ≤ 100,000)이 주어진다. 다음 줄에는 N개의 정수 A[1], A[2], …, A[N]이 주어진다. 다음 줄에는 M(1 ≤ M ≤ 100,000)이 주어진다. 다음 줄에는 M개의 수들이 주어지는데, 이 수들이 A안에 존재하는지 알아내면 된다. 모든 정수의 범위는 -231 보다 크거나 같고 231보다 작다.

### Output)
M개의 줄에 답을 출력한다. 존재하면 1을, 존재하지 않으면 0을 출력한다.

### Source Code - loop)

```python
from sys import stdin
input = stdin.readline

def Binary_Search(li, target):
    start, end = 0, len(li)-1
    
    while start <= end:
        mid = (start + end) // 2

        if li[mid] == target:
            return 1

        elif li[mid] < target:
            start = mid + 1
        else:
            end = mid - 1

    return 0


N = int(input())
num_list = list(map(int, input().split()))
num_list.sort()

M = int(input())
check_list = list(map(int, input().split()))

for target in check_list:
    print(Binary_Search(num_list, target))
```

단순 Binary Search에 관련된 문제이다.



첫 번째 그림은 Binary Search를 이용하였는데 3번의 step 만에 결과를 찾아내지만
두 번째 그림과 같이 처음부터 찾는다면 11번의 step을 거쳐 결과를 찾아낸다.

- 이진 탐색 알고리즘은 정렬되어 있는 리스트에서 탐색 범위를 절반씩 좁혀가며 데이터를 탐색하는 방법이다.

### Source Code - recursion)

```python
import sys
from sys import stdin
sys.setrecursionlimit(10**9)

input = stdin.readline

def Binary_Search(li, target, start, end):
    if start > end:
        return 0
    mid = (start + end) // 2

    if li[mid] == target:
        return 1
    elif li[mid] > target:
        return Binary_Search(li, target, start, mid-1)
    else:
        return Binary_Search(li, target, mid+1, end)


N = int(input())
num_list = list(map(int, input().split()))
num_list.sort()

M = int(input())
check_list = list(map(int, input().split()))

for target in check_list:
    print(Binary_Search(num_list, target, 0, N-1))
```
