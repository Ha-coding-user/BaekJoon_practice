### Q)
자연수 N과 정수 K가 주어졌을 때 이항 계수 nCk 를 1,000,000,007로 나눈 나머지를 구하는 프로그램을 작성하시오.

### Input)
첫째 줄에 
\(N\)과 
\(K\)가 주어진다. (1 ≤ 
\(N\) ≤ 4,000,000, 0 ≤ 
\(K\) ≤ 
\(N\))

### Output)
nCk를 1,000,000,007로 나눈 나머지를 출력한다.

### Source Code)

```python
import sys
from sys import stdin
input = stdin.readline

N, K = map(int, input().split())
p = 1000000007

def factorial(n):
    res = 1
    for i in range(1, n+1):
        res *= i
        res %= p

    return res

def power(n, k):
    if k == 1:
        return n
    
    x = power(n, k//2)
    if k % 2 == 0:
        return x * x % p
    else:
        return x * x * n % p

top = factorial(N)
bottom = factorial(N-K) * factorial(K) % p
print(top * power(bottom, p-2) % p)
```
