### Q)
자연수 A를 B번 곱한 수를 알고 싶다. 단 구하려는 수가 매우 커질 수 있으므로 이를 C로 나눈 나머지를 구하는 프로그램을 작성하시오.

### Input)
첫째 줄에 A, B, C가 빈 칸을 사이에 두고 순서대로 주어진다. A, B, C는 모두 2,147,483,647 이하의 자연수이다.

### Output)
첫째 줄에 A를 B번 곱한 수를 C로 나눈 나머지를 출력한다.

### Source Code)

```python
from sys import stdin
input = stdin.readline

A, B, C = map(int, input().split())
result = 1

def divide_conquer(A, B, C):
    if B == 1:
        return A % C
    else:
        x = divide_conquer(A, B//2 ,C)
        if B % 2 == 0:
            return (x * x) % C
        else:
            return (x * x * A) % C
        
print(divide_conquer(A, B, C))
```

만약 이 알고리즘을 사용하지 않고 단순 계산을 한다면 시간 복잡도는 O(n)이지만, 분할 정복을 이용해 계산하면 O(logn)의 시간 복잡도로 계산이 가능하다.



간단히 2의 33거듭제곱을 계산하는 과정을 살펴보자.



위 source code는 recursion을 이용해 계산한 것이고, 반복문을 이용해 계산하는 방법은 다음과 같다.

```python
def fpow(C, n):
    res = 1
    while n:
        if n % 2 == 1:
            res *= C
        C *= C
        n //= 2
    return res
```
