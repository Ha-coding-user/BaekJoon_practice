### Q)
한 개의 회의실이 있는데 이를 사용하고자 하는 N개의 회의에 대하여 회의실 사용표를 만들려고 한다. 각 회의 I에 대해 시작시간과 끝나는 시간이 주어져 있고, 각 회의가 겹치지 않게 하면서 회의실을 사용할 수 있는 회의의 최대 개수를 찾아보자. 단, 회의는 한번 시작하면 중간에 중단될 수 없으며 한 회의가 끝나는 것과 동시에 다음 회의가 시작될 수 있다. 회의의 시작시간과 끝나는 시간이 같을 수도 있다. 이 경우에는 시작하자마자 끝나는 것으로 생각하면 된다.

### Input)
첫째 줄에 회의의 수 N(1 ≤ N ≤ 100,000)이 주어진다. 둘째 줄부터 N+1 줄까지 각 회의의 정보가 주어지는데 이것은 공백을 사이에 두고 회의의 시작시간과 끝나는 시간이 주어진다. 시작 시간과 끝나는 시간은 231-1보다 작거나 같은 자연수 또는 0이다.

### Output)
첫째 줄에 최대 사용할 수 있는 회의의 최대 개수를 출력한다.

### Source Code)

```python
from sys import stdin
input = stdin.readline

N = int(input())
timeline = [[0]*2 for _ in range(N)]

for i in range(N):
    timeline[i][0], timeline[i][1] = map(int, input().split())

timeline.sort(key = lambda x : (x[1], x[0])) 

end = timeline[0][1]
count = 1

for i in range(1, N):
    if timeline[i][0] >= end:
        count += 1
        end = timeline[i][1]

print(count)
```

이 문제의 경우 시작 시간과 끝나는 시간이 같은 회의에 대해서 고려하는게 어려웠다.

다음 예를 통해 확인해보자

만약 회의가 3개 있는데 (1,9), (8,9), (9,9) 이렇게 있다고 가정해보자.
여기서 가능한 경우는 다음과 같다.
1. (1, 9) / (9, 9)
2. (8, 9) / (9, 9)
즉, 2 경우에서 모두 (9, 9)를 고려해주어야 한다.
따라서 이를 위해 시작시간과 끝 시간에 대해 정렬을 해주어야 한다.
