# 2565 전깃줄

### Q)
두 전봇대 A와 B 사이에 하나 둘씩 전깃줄을 추가하다 보니 전깃줄이 서로 교차하는 경우가 발생하였다. 합선의 위험이 있어 이들 중 몇 개의 전깃줄을 없애 전깃줄이 교차하지 않도록 만들려고 한다.

예를 들어, < 그림 1 >과 같이 전깃줄이 연결되어 있는 경우 A의 1번 위치와 B의 8번 위치를 잇는 전깃줄, A의 3번 위치와 B의 9번 위치를 잇는 전깃줄, A의 4번 위치와 B의 1번 위치를 잇는 전깃줄을 없애면 남아있는 모든 전깃줄이 서로 교차하지 않게 된다.



전깃줄이 전봇대에 연결되는 위치는 전봇대 위에서부터 차례대로 번호가 매겨진다. 전깃줄의 개수와 전깃줄들이 두 전봇대에 연결되는 위치의 번호가 주어질 때, 남아있는 모든 전깃줄이 서로 교차하지 않게 하기 위해 없애야 하는 전깃줄의 최소 개수를 구하는 프로그램을 작성하시오.

### Input)
전깃줄이 전봇대에 연결되는 위치는 전봇대 위에서부터 차례대로 번호가 매겨진다. 전깃줄의 개수와 전깃줄들이 두 전봇대에 연결되는 위치의 번호가 주어질 때, 남아있는 모든 전깃줄이 서로 교차하지 않게 하기 위해 없애야 하는 전깃줄의 최소 개수를 구하는 프로그램을 작성하시오.

### Output)
첫째 줄에 남아있는 모든 전깃줄이 서로 교차하지 않게 하기 위해 없애야 하는 전깃줄의 최소 개수를 출력한다.

### Source Code)

```python
from sys import stdin
input = stdin.readline

n = int(input())

line = [0] * 501 # i번째 idx에 해당 숫자를 넣기 위해 크기 1늘려서 생성
dp = [1] * 501
max_idx = 0 # 최대 idx만 체크하고 거기까지만 확인하기 위해

# A는 idx라 생각하면 되고 각 idx별로 B에 연결된 번호를 저장
for i in range(n):
    idx, num = map(int, input().split())
    line[idx] = num
    if idx > max_idx:
        max_idx = idx

# 그냥 line에 저장된 수열의 가장 긴 부분 수열을 찾으면 됨
for i in range(2, max_idx+1):
    for j in range(1, i):
        # 다만 line[j]가 0인 경우는 넘겨줘야 수가 추가가 안됨
        if line[j] == 0:
            continue
        if line[i] > line[j]:
            dp[i] = max(dp[i], dp[j]+1)

print(n-max(dp))
```

LIS 알고리즘 문제와 동일하다.

A의 위치는 line 리스트의 index로 생각한다.
B의 위치는 line[idx]의 값인데 여기서 idx는 B의 위치와 연결된 A의 위치를 의미한다.

즉, line = [0, 0, 0, 0, 0, ...]로 초기화했을 때 입력으로 1 8 을 받았다면
line = [0, 8, 0, 0, 0, ...] 이 되는 것이다.
모든 입력을 받아 line 리스트에 저장한 후 line리스트에서 0을 포함하지 않은 최장 부분 수열을 생성하면 된다.
