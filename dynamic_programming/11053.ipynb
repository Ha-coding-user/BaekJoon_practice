# 11053 가장 긴 증가하는 부분 수열 ★

### Q)
수열 A가 주어졌을 때, 가장 긴 증가하는 부분 수열을 구하는 프로그램을 작성하시오.

예를 들어, 수열 A = {10, 20, 10, 30, 20, 50} 인 경우에 가장 긴 증가하는 부분 수열은 A = {10, 20, 10, 30, 20, 50} 이고, 길이는 4이다.

### Input)
첫째 줄에 수열 A의 크기 N (1 ≤ N ≤ 1,000)이 주어진다.

둘째 줄에는 수열 A를 이루고 있는 Ai가 주어진다. (1 ≤ Ai ≤ 1,000)

### Output)
첫째 줄에 수열 A의 가장 긴 증가하는 부분 수열의 길이를 출력한다.

### Source Code)

```python
from sys import stdin
input = stdin.readline

n = int(input())
A = list(map(int, input().split()))
dp = [1 for _ in range(n)]

for i in range(1, n):
    temp = []
    for j in range(i):
        if A[i] > A[j]:
            temp.append(dp[j])

    if temp:
        dp[i] = max(temp) + 1

'''
위 코드는 다음과 같이 바꿀 수도 있다

for i in range(1, n):
    for j in range(i):
        if A[i] > A[j]:
            dp[i] = max(dp[i], dp[j] + 1)
'''

print(max(dp))
```

dp[i] : A[i]를 마지막 값으로 가지는 가장 긴 증가부분수열의 길이
- A[i]가 어떤 증가부분수열의 마지막 값이 되려면 A[i] 값이 추가되기 전 증가부분수열의 마지막 값이 A[i]보다 작은 값이여야 한다.
- A[i]가 추가될 수 있는 증가부분수열 중 가장 긴 수열의 길이에 1을 더한 값

A 배열에서 A[i] > A[j] : True 이면 temp 리스트에 추가하고 마지막에 A[i] 에 max(temp) 값을 넣으면 된다.
